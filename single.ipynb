```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.models import load_model
from keras.utils import load_img, img_to_array
import cv2
import matplotlib.pyplot as plt
```

```python
# load model
model1 = load_model("./models/vgg_malaria.h5")
model2 = load_model("./models/inception_resnet_malaria.h5")
```

```python
img = cv2.imread('./images/first_test.jpg')

plt.imshow(img)
plt.show()
```

```python
# BGR2RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#Resizing it to (64,64,3)
img = cv2.resize(img,(132,132))

#Expand dims in axis = 0
img = np.expand_dims(img, axis=0)
```

```python
# Generate prediction for vgg-19 model
pred = model1.predict(img)

if(pred >0.5):
    pred = 'Uninfected'
else:
    pred = 'Parasitized'
    
print(pred)
```

```python
# Generate prediction for inception resnet model
pred = model2.predict(img)

if(pred >0.5):
    pred = 'Uninfected'
else:
    pred = 'Parasitized'
    
print(pred)
```
